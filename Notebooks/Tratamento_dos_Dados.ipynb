```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
```

### **Funções auxiliares**

```python
def adiciona_coluna_estado_e_codigo(dataframe,colname):
  dataframe.drop(27, inplace=True)

  nomes_estados = []
  codigos_estados = []

  for uf in dataframe[colname]:
    if len(uf.split(" ")) == 2:
      nomes_estados.append(uf.split(" ")[1])
    else:
      nomes_estados.append(" ".join(uf.split(" ")[1:]))

    codigos_estados.append(uf.split(" ")[0])

  dataframe.drop([colname], axis=1, inplace=True)
  dataframe.insert(loc=0, column='Estado', value=nomes_estados)
  dataframe.insert(loc=1, column='Código do Estado', value=codigos_estados)

def remove_colunas(dataframe):
    if {'2021'}.issubset(dataframe.columns):
      dataframe.drop(["2021"], axis=1, inplace=True)

def converte_coluna_de_anos_em_inteiro(dataframe):
  inicio = 2001
  final = 2020
  for ano in range(inicio, final + 1):
    if {str(ano)}.issubset(dataframe.columns):
      dataframe[str(ano)] = dataframe[str(ano)].astype(int)
```

### **Tratamento dos dados de Tuberculose**


```python
# Incidência de tuberculose ao longo dos anos
url_raw = "https://raw.githubusercontent.com/NadiaaOliverr/O-Brasil-em-Dados-CCF-425/main/Dados/Brutos/Casos-Tuberculose.csv?token=AHQOTIRMXDJ77C5UWWW6OULBI6TYG"
df_casos_tuberculose = pd.read_csv(url_raw, encoding="utf-8", delimiter=";")
adiciona_coluna_estado_e_codigo(df_casos_tuberculose,"UF de notificação")
remove_colunas(df_casos_tuberculose)
converte_coluna_de_anos_em_inteiro(df_casos_tuberculose)
df_casos_tuberculose.to_csv("casos_tuberculose.csv", encoding="utf-8", sep=";", index=False)
df_casos_tuberculose.head(n=10)
```
